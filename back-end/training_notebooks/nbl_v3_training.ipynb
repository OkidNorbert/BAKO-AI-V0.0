# Training

```python
from ultralytics import YOLO

model = YOLO('yolo11n.pt')

results = model.train(
    data='/content/NBL-3/data.yaml',
    epochs=100,
    batch=16,
    imgsz=640,
    patience=100,
    device=0,
    workers=8,
    project='nbl_v3_runs',
    name='train',
    exist_ok=False,
    pretrained=True,
    optimizer='auto',
    verbose=True,
    seed=0,
    deterministic=True,
    val=True,
    plots=True
)
```
